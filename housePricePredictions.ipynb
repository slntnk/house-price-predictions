# ÇOKLU LİNEER REGRESYON

Bir bağımlı değişkenin birden fazla bağımsız değişken tarafından tahmin
edildiği regresyon modelidir.

Genel olarak ifade edildiği denklem:
Y= β0 + β1X1 + β2X2 + ...+ βpXp + ϵ

Y: bağımlı değişken (hedef değişken)

β0: Sabit katsayı

β1,β2,...,βp: Bağımsız değişken katsayıları (her bir bağımsız değişkenin bağımlı değişken üzerindeki etkisini belirler. Genelde en küçük kareler yöntemi kullanılarak hesaplanır.

X1,X2,...,X: Bağımlı değişkenler

ϵ: Hata terimi


```python
#Çoklu Doğrusal Regresyon Modeli
#model eğitimi
from sklearn.linear_model import LinearRegression
model_lr =LinearRegression()
#eğitilen modele verileri uygulama
model_lr.fit(x_train,y_train)
```

```python
model_lr.intercept_
```

```python
model_lr.coef_
```

```python
#denklem ile gösterilmesi

katsayilar = model_lr.coef_
sabit = model_lr.intercept_

denklem = f"Y = {model_lr.intercept_}"
for i, katsayi in enumerate(katsayilar):
    denklem += f" + {katsayi:.2f} * {x_test.columns[i]}"

print("Oluşturulan Denklem:", denklem)

ornek_veri = x_test.iloc[4]
tahmin_edilen_deger = sabit
for i, katsayi in enumerate(katsayilar):
    tahmin_edilen_deger += katsayi * ornek_veri.iloc[i]

print("Tahmin edilen değer:", tahmin_edilen_deger)
```

```python
#eğitilen model üzerinden test verisiyle tahmin yapılması
print("Tahmin edilen değer : " + str(model_lr.predict([x_test.iloc[4]])))
print("Gerçek değer : " + str(y_test.iloc[4]))
```

```python
print("Lineer Regresyon Modeli Doğruluk oranı : % {:.3f} ".format(model_lr.score(x_test, y_test)* 100))
```

# KARAR AĞACI REGRESYONU
Denetimli yöntemlerden biridir. Veri seti içerisindeki karmaşık yapıları basit karar yapılarına dönüştürmektedir. Heterojen veri setleri belirlenmiş bir hedef değişkene göre homojen alt gruplara ayrılır ve hedef değişken tahmin edilmeye çalışılır.

 Her iç düğümde en iyi bölme noktasını belirlemek için bir bölme kriteri kullanır. Sınıflandırma problemleri için genellikle entropy, regresyon problemleri için ortalama kare hata kullanılır. CART algoritması en çok kullanılan karar ağacı regresyon algoritmalarından biridir.



```python
# KARAR AĞACI
from sklearn.tree import DecisionTreeRegressor
model_dtr2 = DecisionTreeRegressor(random_state = 10,max_depth=3)
model_dtr2.fit(x_train, y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15, 8))
plot_tree(model_dtr2, filled=True)
plt.show()
```

```python
from sklearn.tree import export_text
tree_rules = export_text(model_dtr2, feature_names=list(x_train.columns))
print(tree_rules)
```

```python
print("Doğruluk oranı : %",model_dtr2.score(x_test, y_test)* 100)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_dtr = DecisionTreeRegressor(random_state = 100)
model_dtr.fit(x_train, y_train)
```

```python
print("Karar Ağacının Maksimum Derinliği:", model_dtr.get_depth())
print("Karar Ağacının Toplam Düğüm Sayısı:",model_dtr.tree_.node_count)
```

```python
plt.figure(figsize=(18,8))
plot_tree(model_dtr,filled=True)
plt.show()
```

```python
print("Tahmin edilen değer : " + str(model_dtr.predict([x_test.iloc[4]])))
print("Gerçek değer : " + str(y_test.iloc[4]))
```

```python
print("Karar Ağacı Modeli Doğruluk oranı : % {:.3f} ".format(model_dtr.score(x_test, y_test)* 100))
```

# Random Forest Yöntemi
Temeli birçok karar ağacının ürettiği tahminlerin bir araya getirilerek değerlendirilmesine dayanır. Ağaç modelleri oluşturulurken gözlemler ve değişkenler rastgele seçilerek çeşitlilik arttırılır.

 Oluşturulan ağaç modellerinin tahminleri sınıflandırma yöntemlerinde en çok tekrar edilen tahmin alınırken regresyon yöntemlerinde tahminlerin ortalaması alınarak yapılır.


```python
#RANDOM FOREST
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(x_train, y_train)
```

```python
print("Ağaç Sayısı:",len(model_rf.estimators_))
```

```python
önem_sırası = model_rf.feature_importances_
feature_names = x_train.columns
sıralama = np.argsort(önem_sırası)[::-1]

print("Özellik Önem Sıralaması:")
for f in range(x_train.shape[1]):
    print("%d. Özellik: %s (%f)" % (f + 1, feature_names[sıralama[f]], önem_sırası[sıralama[f]]))
```

```python
print("Tahmin edilen değer : " + str(model_rf.predict([x_test.iloc[4]])))
print("Gerçek değer : " + str(y_test.iloc[4]))
```

```python
print("Lineer Regresyon Modeli Doğruluk oranı : % {:.3f} ".format(model_lr.score(x_test, y_test)* 100))
print("Karar Ağacı Modeli Doğruluk oranı : % {:.3f} ".format(model_dtr.score(x_test, y_test)* 100))
print("Random Forest Modeli Doğruluk oranı : % {:.3f} ".format(model_rf.score(x_test, y_test)* 100))
```

```python
from sklearn.metrics import mean_absolute_error

mae_lr = mean_absolute_error(y_test, model_lr.predict(x_test))
mae_dtr = mean_absolute_error(y_test, model_dtr.predict(x_test))
mae_rf = mean_absolute_error(y_test, model_rf.predict(x_test))

print("Lineer Regresyon Modeli MAE: {:.3f}".format(mae_lr))
print("Karar Ağacı Modeli MAE: {:.3f}".format(mae_dtr))
print("Random Forest Modeli MAE: {:.3f}".format(mae_rf))
```

```python
# Gerçek ve tahmin edilen değerlerin scatter plotunu oluşturalım
plt.figure(figsize=(10, 6))

plt.scatter(y_test, model_lr.predict(x_test), color='purple', label='Lineer Regresyon',s=10)
plt.scatter(y_test, model_dtr.predict(x_test), color='red', label='Karar Ağacı',s=10)
plt.scatter(y_test, model_rf.predict(x_test), color='green', label='Random Forest',s=10)

plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.title('Gerçek ve Tahmin Edilen Değerlerin Karşılaştırılması')

plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='black')
plt.legend()
plt.show()
```
